```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
import datetime

ticker = "BAC"

data = yf.download(ticker,start="2023-01-01", end="2024-01-01", interval='1h')
```

```python
from statsmodels.tsa.stattools import adfuller

def test_stationarity(timeseries):
    print('Rsultats du Test de Dickey-Fuller :')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic', 'p-value', '#Lags Used', 'Number of Observations Used'])
    for key, value in dftest[4].items():
        dfoutput['Critical Value (%s)' % key] = value
    print(dfoutput)

test_stationarity(data['Close'])
```

```python
data_diff = data['Close'].diff()
data_diff.dropna(inplace=True)
```

```python
test_stationarity(data_diff)
```

```python
import matplotlib.pyplot as plt

data_diff = data['Close'].diff().dropna()

plt.figure(figsize=(15, 6))

plt.subplot(1,2,1)
plt.plot(data['Close'])
plt.title('Donnes Originales (Close Price)')
plt.xlabel('Date')
plt.ylabel('Prix de Clture')
plt.xticks(rotation=45)

plt.subplot(1,2,2)
plt.plot(data_diff)
plt.title('Donnes Diffrencies (1st Difference)')
plt.xlabel('Date')
plt.ylabel('Diffrence de Prix')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
train = data_diff[:int(0.9*len(data))]
test = data_diff[int(0.9*len(data)):]
```

```python
del data_diff
```

```python
train.shape
```

```python
from pmdarima import auto_arima

model = auto_arima(train, start_p=1, start_q=1,
                   max_p=5, max_q=5, m=48,  
                   seasonal=True, 
                   d=1, D=1, trace=True,
                   error_action='ignore',  
                   suppress_warnings=True, 
                   stepwise=True)

print(model.summary())
```

```python
model_fit = model.fit(train)

print(model_fit.summary())
```

```python
model_fit.plot_diagnostics(figsize=(15, 8))
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

predictions = model.predict(n_periods=len(test))
mse = mean_squared_error(test, predictions)
print(f"Mean Squared Error: {mse}")
```

```python
plt.figure(figsize=(12,6))
plt.plot(train[-168:].index, train[-168:], label='Train')
plt.plot(test.index, test, label='Test')
plt.plot(test.index, predictions, label='ARIMA Predictions')
plt.legend()
plt.show()
```
